## Task 1  
Start your notebook and import required libraries.

**Goal**  
Prepare the environment to work with images.


```python
# Task 1
# Import the required libraries
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Print a message to confirm the imports worked
print("Libraries imported successfully.")
```

## Task 2  
Download the image from the provided URL.

**Goal**  
Save the image into the Colab runtime so it can be read by OpenCV.


```python
# Task 2
# Import urllib.request
import urllib.request

# Store the image URL in a variable named url
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/fruits.jpg"

# Store the file name as fruits.jpg
filename = "fruits.jpg"

# Download the image
urllib.request.urlretrieve(url, filename)

# Print a confirmation message
print("Image downloaded successfully.")
```

## Task 3  
Read the image using OpenCV.

**Goal**  
Load the downloaded image into a variable.


```python
# Task 3
# Read the image using cv2.imread
image_bgr = cv2.imread(filename)

# Check if the image loaded correctly
if image_bgr is not None:
    print("Image loaded successfully.")
else:
    print("Failed to load image.")
```

## Task 4  
Inspect the image type and shape.

**Goal**  
Understand how the image is stored as data.


```python
# Task 4
# Print the type of the image
print("Type of the image:", type(image_bgr))

# Print the shape of the image
print("Shape of the image:", image_bgr.shape)
```

## Task 5  
Convert the image from BGR to RGB.

**Goal**  
Prepare the image for correct display using Matplotlib.


```python
# Task 5
# Convert image_bgr to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

## Task 6  
Display the RGB image.

**Goal**  
Visualize the image correctly.


```python
# Task 6
# Display the RGB image using matplotlib
plt.imshow(image_rgb)
# Turn off the axes
plt.axis('off')
# Add a title
plt.title('RGB Image')
plt.savefig('rgb_image.png')
print('<Figure size')
```

## Task 7  
Display RGB channels separately.

**Goal**  
Understand how a color image is made of three channels.


```python
# Task 7
# Extract Red, Green, and Blue channels from image_rgb
red_channel = image_rgb[:, :, 0]
green_channel = image_rgb[:, :, 1]
blue_channel = image_rgb[:, :, 2]

# Display each channel using matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

plt.savefig('rgb_channels.png')
print('<Figure size')
```
